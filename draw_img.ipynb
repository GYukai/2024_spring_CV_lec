```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")
```

```python
ab_aug_only = pd.read_csv("ab_aug_only.csv")
ab_base = pd.read_csv("ab_base.csv")
ab_trans_only = pd.read_csv("ab_trans_only.csv")
final = pd.read_csv("final.csv")
```

```python
final
```

```python
steps = []
a = range(300, 5000, 300)
for i in a:
    steps.append(i)
steps.append(5000)
```

```python
ab_aug_only["bbox/AP"][:17]
```

```python
plt.figure(figsize=(10, 5), dpi=300)
plt.plot(steps, ab_aug_only["bbox/AP"][:17], label="Aug only")
plt.plot(steps, ab_base["bbox/AP"][:17], label="Baseline")
plt.plot(steps, ab_trans_only["bbox/AP"][:17], label="Trans only")
plt.plot(steps, final["bbox/AP"][:17], label="Final")
plt.xlabel("Steps")
plt.ylabel("AP")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi=300)
plt.plot(steps, ab_aug_only["bbox/AP50"][:17], label="Aug only")
plt.plot(steps, ab_base["bbox/AP50"][:17], label="Baseline")
plt.plot(steps, ab_trans_only["bbox/AP50"][:17], label="Trans only")
plt.plot(steps, final["bbox/AP50"][:17], label="Final")
plt.xlabel("Steps")
plt.ylabel("AP50")
plt.legend()
```
